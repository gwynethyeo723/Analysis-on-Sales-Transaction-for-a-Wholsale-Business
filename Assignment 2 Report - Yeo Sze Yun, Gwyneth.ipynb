## 2.3 Preparation of data<a id="howprep"></a>

For easy reference when gathering the data to make the visualisations, I changed the column names as shown below. 

```python
df = df.rename(columns={"OrderID":"id", "Order Year":"year","Order Quarter": "quarter","Order Month":"month",
                        "Order Day":"day","Customer Name":"cust_name","Customer Country":"cust_country",
                        "Customer City":"cust_city","ProductName":"p_name","Product CategoryName":"p_catname",
                        "Quantity":"quantity","UnitPrice in USD":"u_price","Unit Discount in %":"u_discount",
                        "UnitCost in USD":"u_cost","Shipper Name":"ship_name",
                        "Freight Cost (Paid by Customer)": "freight_cost"})
```

I had also included additional columns such as "total_price" which calculates and stores the total price after discount for each product in each order and round it off to 2 decimal place. This helps to compute the revenue later on. 

The discounted unit price multiplied by the quantity ordered will give the total price of each product in each order ordered. 

```python
df["total_price"]=(1-df["u_discount"])*df["u_price"]*df["quantity"]
df["total_price"]=df["total_price"].round(2)
```

Similarly, I had added the total profit earned for each product in each order and rounded it off to 2 decimal place for easy computation of the profits earned later on. 

The total discounted price subtracted by the unit cost multiplied by the quantity ordered will give the profit.

```python
df["profit"]=df["total_price"]-(df["u_cost"]*df["quantity"])
df["profit"]=df["profit"].round(2)
```

In order to sort and display the months in chronological order when making the visualisations, I encoded each month to their specific numerics for easy usage. 

```python
replace_month = {'month':{'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,"September":9,
                          'October':10,'November':11,'December':12}}
df.replace(replace_month, inplace=True)
```

To make the visualisations neater, I rounded off 'u_price','u_discount','u_cost' and 'freight_cost' to 2 decimal place. 

```python
df['u_price']=df['u_price'].round(2)
df['u_discount']=df['u_discount'].round(2)
df['u_cost']=df['u_cost'].round(2)
df['freight_cost']=df['freight_cost'].round(2)
```

This is how the dataset that I will be using to make visualisations look like.

# 3. Visualisations<a id="vis"></a>

## 3.1 Revenue And Profits (2014 to 2018)<a id="v1"></a>

```python
yearly_revenue_v1 = df.groupby('year')['total_price'].sum() #Find the revenue earned per year
yearly_profits_v1 = df.groupby('year')['profit'].sum() # Find the profits earned per year
yearly_profits_v1 = ((yearly_profits_v1/yearly_revenue_v1)*100).round(2) #Find the profit margin
year_v1 = yearly_revenue_v1.index.astype(str) #Put the year index to string so that it does not show as decimal in the visualisation

fig, v1_b = plt.subplots(figsize=(15,5)) # Create a figure and subplot for revenue chart 

# Plot revenue chart
v1_b.bar(year_v1, yearly_revenue_v1, width=0.35, label='Revenue', color='#AFE4FE')

# State title of plot and the axes labels
v1_b.set_title('Revenue And Profits (2014 to 2018)', fontsize = 20) 
v1_b.set_ylabel('Revenue ($)', fontsize = 15) 
v1_b.set_xlabel('Year', fontsize = 15)

v1_b.grid(c = 'g', alpha = .4,linestyle = '-.') #Set grid lines for easier viewing of data 
v1_b.set_ylim(bottom = 0, top = 800000)# Set ranges for the y-axis 

v1_b.tick_params(labelsize = 12, axis = 'both') # Set font size of ticks on the x-axis and left y-axis 

for x,y in zip(year_v1 ,yearly_revenue_v1): # Display data labels

    label = y
    
    plt.annotate(label,
                 (x,y),
                 textcoords="offset points",
                 xytext=(0,0),
                 fontsize = 12,
                 ha='center') 
    
v1_l = v1_b.twinx() # Instantiate a second axes that shares the same x-axis

# Plot profit margins chart
v1_l.plot(year_v1, yearly_profits_v1, label = 'Profits', color= '#E5A4BE',marker = 'o')

# Set right y-axis label 
v1_l.set_ylabel('Profit Margin in Percentage (%)', fontsize = 15)

 # Display data labels
for x,y in zip(year_v1 ,yearly_profits_v1):

    label = y

    plt.annotate(label,
                 (x,y),
                 textcoords="offset points", 
                 xytext=(8,3),
                 fontsize = 12,
                 ha='center') 

# Set font size of ticks on the right y-axis
v1_l.tick_params(labelsize = 12, axis = 'y')

# Set appropriate position for the legend
fig.legend(loc=(0.88,0.82)) 

# Set tight layout for figure
fig.tight_layout()

# Display plot
plt.show()
```

#### Question it answers: 
- What is the trend of revenue from 2014 to 2018?
- What is the trend of profit from 2014 to 2018? 

#### Core findings and insights:
The visualisation illustrates the revenue and profits from 2014 to 2018. It can been seen that the revenue was steadily increasing from 2432.12 dollars in 2014 to 734547.05 dollars in 2017 but suffered a drastic drop to 302164.63 dollars in 2018. Based on the trending, the profits hover between 10.79% to 17.97%. The visulisation also shows that high revenue does not equate to high profits. 2014 is the most profitable year for the company with 17.97% profits. However, since the dataset for 2014 is not complete, 2016 would be the next most profitable year for the company with 14.01% profits. ## 3.3 Average Order Price (2014 to 2018) <a id="v3"></a>

```python
year_v3 = df['year'].unique().astype(str) #Find unique years and set to string 
year_v3 = [year_v3[3],year_v3[4],year_v3[2],year_v3[1],year_v3[0]] #Order year by ascending order of year

avg_orderList_v3 = []

for i in year_v3: #For every year, 
    yeardata_v3 = df[df['year']==int(i)] # Filter rows from df where 'year'==int(i) and store them in a DataFrame 
    yearly_revenue_v3 = yeardata_v3['total_price'].sum() #Find the revenue for that year
    unique_ids_v3 = len(yeardata_v3['id'].unique()) #Find how many orders there were for that year
    avg_orderprice_v3 = (yearly_revenue_v3/unique_ids_v3).round(2) #Find the average order price
    avg_orderList_v3.append(avg_orderprice_v3) #Append to list

# Create a figure and subplot
fig, v3 = plt.subplots(figsize=(8, 6) )

# Create plot containing line chart for average order price
v3.plot(year_v3, avg_orderList_v3, color= '#8ECD9A', marker ='o')

# Set font size of ticks on the x-axis and y-axis
v3.tick_params(labelsize = 12,axis = 'both') 

# Set ranges for the x-axis and y-axis
v3.set_xlim(left = -1 , right = 5)
v3.set_ylim(bottom = 0, top = 2700)

# State title of plot and the axes labels
v3.set_xlabel('Year', fontsize = 15)
v3.set_ylabel('Average Order Price ($)', fontsize = 15)
v3.set_title('Average Order Price (2014 to 2018)', fontsize = 20)

# Display data labels
for x, y in zip(year_v3,avg_orderList_v3): 

    label = y 

    v3.annotate(label, 
                 (x,y),
                 textcoords="offset points", 
                 xytext=(35,-12),
                 fontsize = 11,
                 ha='center') 

# Set tight layout for figure
fig.tight_layout()

# Display plot
plt.show()
```

## 3.6 Distribution Of Revenue By Customer (2014 to 2018) <a id = "v6"></a>

```python
total_revenue_v6 = df['total_price'].sum() #Find the total revenue for 2014 to 2018
cust_revenue_v6 = df.groupby('cust_name')['total_price'].sum().round(2) #Find the revenue each customer contributed
cust_revenue_v6.sort_values(ascending=True, inplace=True) #Sort the values in ascending order of revenue

bottom_cust_revenue_v6 = cust_revenue_v6[:5].sum().round(2) #Find the top 5 customers that contributed the least revenue

top_cust_revenue_v6 = cust_revenue_v6[-5:].sum().round(2) #Find the top 5 customers that contributed the most to the revenue

remaining_cust_revenue_v6 = total_revenue_v6 - bottom_cust_revenue_v6 - top_cust_revenue_v6 #Find the remaining amount of revenue

revenue_piechart_v6 = [remaining_cust_revenue_v6, bottom_cust_revenue_v6, top_cust_revenue_v6]
colour = ['#AFE4FE','#A4C3D2','#78A2CC'] #Set colours 
labels_v6 = ['Rest','Bottom 5','Top 5'] #Set labels 

# Create a figure and subplot
fig, v6 = plt.subplots(figsize = (8,6))

# Create Customzied Pie Chart
v6.pie(revenue_piechart_v6,explode = (0.1,0.1,0.1), labels = labels_v6, autopct = '%1.2f%%', textprops = {'fontsize' : 15}, startangle=90, colors = colour)

# Set title of plot
v6.set_title('Distribution Of Revenue By Customer (2014 to 2018)', fontsize = 20)

# Equal aspect ratio ensures that pie is drawn as a circle
v6.axis('equal')

# Set tight layout for figure
fig.tight_layout()

# Display plot
plt.show()
```
